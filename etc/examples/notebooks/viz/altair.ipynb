```python
import altair as alt
import pandas as pd
```

```python
# alt.enable_mime_rendering()
cars = alt.load_dataset('cars')
cars
```

```python
chart = alt.Chart(cars).mark_circle().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
)
chart 
```

```python
from vega_datasets import data
iris = data.iris()
iris
```

```python
alt.Chart(iris).mark_point().encode(
    x='petalLength',
    y='petalWidth',
    color='species'
)
```
